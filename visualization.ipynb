# Visualization

```python
import pandas as pd
# import plotly.express as px
# import plotly.graph_objects as go


katalog = pd.read_csv('katalog_gempa.tsv', sep='\t')
```

```python
katalog.fillna('')
```

```python
# Distribusi magnitudo gempa
katalog.groupby('mag').count()['remark'].plot(figsize=(16, 4));
```

```python
# Frekuensi banyak gempa berdasarkan waktu
katalog.groupby('tgl').count()['remark'].plot(figsize=(16, 4));
```

```python
# Mean gempa berdasarkan waktu
katalog.groupby('tgl').mean()['mag'].plot(figsize=(16, 4));
```
